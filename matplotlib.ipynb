```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(12,22)
```

## A Simple Plot

```python
#plotting using matplotlib
#plt scatter
plt.scatter(x,y,c="g")         
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Simple Plot")
plt.show()
```

## To Save the result

```python
## to save the result of the diagram
plt.savefig("temp.png")
```

## y =X**2   graph

```python
y=x**2
plt.plot(x,y,"r--")
```

##  Subplots

```python
plt.subplot(2,3,1)
plt.plot(x,y,"r")
plt.subplot(2,3,2)      #in subplot  2 is no of rows and 3 is no of columns
plt.plot(x,y,"b")       #1,2,3...  position 
plt.subplot(2,3,3)
plt.plot(x,y,"g")
```

## y=Sin(x)   Graph

```python
np.pi
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y,"r--")
plt.axhline(y=0,color="black",linestyle="-")
```

## Bar plot using Matplotlib

```python
x=[3,7,11]
y=[2,5,8]

a=[15,5,10]
b=[10,10,10]

plt.bar(x,y)

plt.bar(a,b,color="g")
plt.xlabel="X axis"
plt.ylabel="Y axis"
plt.title("Bar plot")
plt.show()
```

## Histograms

```python
a=np.array([12,10,4,34,76,85,30,25,11,5])
plt.hist(a,bins=10)
plt.show()
```

## Box plot

```python
data=[np.random.normal(0,i,100) for i in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
```

## Pie Chart

```python
labels="python","c++","ruby","java"
colors=["gold","red","blue","lightcoral"]
sizes=[100,200,300,400]
explode=(0.1,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,shadow=True,autopct="%1.1f%%")
plt.axis("equal")
plt.show()
```
